```python
df = pd.read_csv('mumbai_restaurants.csv')
```

```python
df['query'] = df['name'] + ' ' + df['address']
```

```python
def get_lat_lng(query):
    url = "https://geocode.search.hereapi.com/v1/geocode"
    params = {
        "q": query,
        "apiKey": HERE_API_KEY
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        items = response.json().get('items')
        if items:
            position = items[0].get('position')
            print(position)
            return position.get('lat'), position.get('lng')
    return None, None
```

```python
# Apply the function and store results
df['latitude'], df['longitude'] = zip(*df['query'].apply(lambda x: get_lat_lng(x)))

# Optional: Save the updated dataframe
df.to_csv('updated_dataset_with_coords.csv', index=False)

print("Latitude and longitude added.")
```
